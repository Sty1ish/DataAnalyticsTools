### 범주형 데이터 사용 예제
- p를 모르는 상황
- 우리 전체 모집단은 100,000명이라고 가정
- 신뢰구간 0.95, 허용 오차 0.01 (1%) 가정

```python
sample_size_proportion(
    p=0.5, error=0.01, confidence_level=0.95, population_size=200000)
```

### 수치형 데이터 사용 예제
- 시그마를 10으로 가정함 (모분산이건 표본분산이건 상관없고, 사전 실험으로 얻기만 하면 무방)
- 우리 전체 모집단은 100,000명이라고 가정
- 신뢰구간 0.95, 허용 오차 0.01 (1%) 가정

```python
sample_size_mean(
    sigma=10, error=1, confidence_level=0.95, population_size=100000)
```
